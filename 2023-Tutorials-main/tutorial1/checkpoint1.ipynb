Let's compare Michigan to Ohio. Which state has the higher average outcome score across all years?

```python
michigan = df[df["STATE"] == "MICHIGAN"].dropna()
ohio = df[df["STATE"] == "OHIO"].dropna()

michigan_avg = michigan["AVG_MATH_8_SCORE"].mean()
ohio_avg = ohio["AVG_MATH_8_SCORE"].mean()

if (michigan_avg > ohio_avg):
    print(f"Michigan has the higher average outcome score across all years by a score of {(michigan_avg - ohio_avg):.1f}")
else:
    print(f"Ohio has the higher average outcome score across all years by a score of {(ohio_avg - michigan_avg):.1f}")
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['GRADES_8_G','AVG_READING_8_SCORE', 'INSTRUCTION_EXPENDITURE_PER_8_STUDENT']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```
